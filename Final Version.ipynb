```python

temp_array = np.zeros([len(data['Percentile Rank']), 2])
```

```python
temp_array[:, 0] = np.floor(data['Percentile Rank'])
```

```python
temp_array[:, 1] = np.array(data['TOT_EMP'])
```

```python
total_employ = np.zeros(101)
```

```python
total_employ
```

```python
Ntemp=temp_array[~np.isnan(temp_array).any(axis=1), :]
```

```python
len(Ntemp)
```

```python
sum(Ntemp[:, 0] ==100)
```

```python
for i in range(len(Ntemp[:, 0])):
    perc = Ntemp[:, 0][i]
    p = int(perc)
    total_employ[p] = total_employ[p] + Ntemp[:, 1][i]
```

```python
total_employ = total_employ / 1000000
```

```python
fig = plt.figure(figsize = (9,6))
plt.scatter(xbin, total_employ)
p = np.polyfit(xbin, total_employ, 2)
plt.plot(xbin,np.polyval(p, xbin))
plt.title('Total Employment by Skill Percentage')
plt.ylabel('Total Employment by million')
plt.xlabel('Skill Percentage as a measure of hourly mean wage')
plt.show()
```

```python
import collections
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def dealstr(s):
    return s.lower().replace('and', '').replace(',', '').split()

data['OCC_TITLE'].map(dealstr).explode().value_counts()[:10]
```

```python
data['title'] = data['OCC_TITLE'].str.lower().replace('and', '').replace(',', '')

titles = list(data['OCC_TITLE'].map(dealstr).explode().value_counts()[:10].index)
titles = [a for a in titles if a not in ['other', 'all', 'except']]
data['title'] = data['title'].apply(
    lambda x:[s for s in titles if s in x][0] 
    if any([s in x for s in titles]) 
    else 'other'
)
```

```python
temp1 = data[data['Percentile Rank'] < 10].groupby(['year', 'title'])['TOT_EMP'].sum().unstack() / 1e7


temp1.drop('other', axis=1).plot(
    kind='bar', figsize=(24, 10)
)
plt.ylabel('Total employments (Unit: 10 million)')
plt.title('Total employments of different job categories in year 2010-2020')
plt.savefig('output.png', dpi=500)
```
